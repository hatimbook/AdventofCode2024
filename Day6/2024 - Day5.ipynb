```python
%%time
with open('input.txt') as f:
    temp = f.read().splitlines()
    temp = [list(x) for x in temp]
h, w = len(temp), len(temp[0])

for c in range(w):
    for r in range(h):
        if temp[r][c] == '^':
            startx, starty = c, r
            temp[r][c] = '.'
            break

inroom = True
currx, curry, currdir = startx, starty, "UP"
locations = set([])

while(inroom):
    inroom, newx, newy, newdir = move(currx, curry, currdir)
    if inroom:
        if (newx,newy) not in locations:
            locations.add((newx,newy))
    currx, curry, currdir = newx, newy, newdir

loop = 0    
for pos, loc in enumerate(locations):
    temp[loc[1]][loc[0]] = '#'
    inroom = True
    currx, curry, currdir = startx, starty, "UP"
    locswdir = set([(currx, curry, currdir)])
   
    while(inroom):
        inroom, newx, newy, newdir = move(currx, curry, currdir)
        if inroom:
            if((newx, newy, newdir) in locswdir):
                loop += 1
                break
            else:
                locswdir.add((newx,newy,newdir))
        currx, curry, currdir = newx, newy, newdir
        
    temp[loc[1]][loc[0]] = '.'
    
print("Part1:", len(locations))
print("Part2:", loop)
```

```python
def move(x,y,d):
    global temp
    if d == 'UP':
        newposx = x
        newposy = y-1
        if newposy > -1:
            if temp[newposy][newposx] == '.':
                return True, newposx, newposy, d
            if temp[newposy][newposx] == '#':
                return True, x, y, "RIGHT"
        else:
            return False, x, y, d
        
    elif d == 'RIGHT':
        newposx = x+1
        newposy = y
        if newposx < w:
            if temp[newposy][newposx] == '.':
                return True, newposx, newposy, d
            if temp[newposy][newposx] == '#':
                return True, x, y, "DOWN"
        else:
            return False, x, y, d
        
    elif d == 'DOWN':
        newposx = x
        newposy = y+1
        if newposy < h:
            if temp[newposy][newposx] == '.':
                return True, newposx, newposy, d
            if temp[newposy][newposx] == '#':
                return True, x, y, "LEFT"
        else:
            return False, x, y, d
        
    elif d == "LEFT":
        newposx = x-1
        newposy = y
        if newposx > -1:
            if temp[newposy][newposx] == '.':
                return True, newposx, newposy, d
            if temp[newposy][newposx] == '#':
                return True, x, y, "UP"
        else:
            return False, x, y, d
     
    return False,x,y,d
    
```

```python
# AI Generated function
# Direction changes dictionary
direction_changes = {
    'UP': 'RIGHT',
    'RIGHT': 'DOWN',
    'DOWN': 'LEFT',
    'LEFT': 'UP'
}
def move(x, y, d):
    """
    Move in a 2D grid based on the given direction.

    Parameters:
    x (int): Current x-coordinate.
    y (int): Current y-coordinate.
    d (str): Direction to move in. Can be 'UP', 'DOWN', 'LEFT', or 'RIGHT'.
    grid (list of list of str): The 2D grid.
    width (int): Width of the grid.
    height (int): Height of the grid.
    direction_changes (dict): Dictionary for direction changes.

    Returns:
    tuple: (bool, int, int, str) indicating if the move was successful, new x, new y, and new direction.
    """
    new_x, new_y = x, y

    if d == 'UP':
        new_y -= 1
    elif d == 'RIGHT':
        new_x += 1
    elif d == 'DOWN':
        new_y += 1
    elif d == 'LEFT':
        new_x -= 1
    else:
        raise ValueError("Invalid direction. Use 'UP', 'DOWN', 'LEFT', or 'RIGHT'.")

    if 0 <= new_x < w and 0 <= new_y < h:  # Check boundaries
        if temp[new_y][new_x] == '.':
            return True, new_x, new_y, d
        elif temp[new_y][new_x] == '#':
            # Change direction on encountering '#'
            return True, x, y, direction_changes[d]
    return False, x, y, d
```
