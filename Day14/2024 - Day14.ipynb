```python
import re
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt

def movepoint(p, v, steps, x_max, y_max):
    newpoint = ((p[0] + v[0]*steps)%x_max, (p[1] + v[1]*steps)%y_max)
    return newpoint

with open('input.txt') as f:
    temp = f.read().splitlines()

new_points = set([])
quad = [0,0,0,0]

x_max, y_max, steps = 101, 103, 100
x_mid, y_mid = x_max //2,  y_max //2

new_points = set([])
for point in temp:
    m = re.match(r'p=(\d+),(\d+) v=(-?\d+),(-?\d+)', point)

    if m:
        p= (int(m.group(1)), int(m.group(2)))
        v = (int(m.group(3)), int(m.group(4)))
        new = movepoint(p, v, steps, x_max, y_max)
        new_points.add(new)
        if new[0] < x_mid and new[1] < y_mid:
            quad[0] += 1
        elif new[0] > x_mid and new[1] < y_mid:
            quad[1] += 1
        elif new[0] < x_mid and new[1] > y_mid:
            quad[2] += 1
        elif new[0] > x_mid and new[1] > y_mid:
            quad[3] += 1      

            print("Part1 = ", np.prod(quad))

x,y = zip(*new_points)

plt.scatter(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Set s')

plt.show()
# Part 2 was done by eyeballing the graphs and finding a recurring pattern at intervals of 101             
```
